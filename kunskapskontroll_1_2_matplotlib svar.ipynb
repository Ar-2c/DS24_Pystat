### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python
# Den objektorienterade metoden nedan enligt exemplet:

x = np.linspace(0, 2, 100) 

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.plot(x, x, label='linear')  
ax.plot(x, x**2, label='quadratic')  
ax.plot(x, x**3, label='cubic')  
ax.set_xlabel('x label')  
ax.set_ylabel('y label')  
ax.set_title("Simple Plot")  
ax.legend()  
```

```python
# Pyplot-metoden enligt nedan:

x = np.linspace(0, 2, 100) 

plt.figure(figsize=(5, 2.7), layout='constrained')
plt.plot(x, x, label='linear')  
plt.plot(x, x**2, label='quadratic')  
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```

```python
# Det som rekommenderas är den Objektorienterade metoden. Särskilt för mer komplicerade plottningar som är tilltänkta att användas i större projekt och OO-metoden ger större frihet att göra specifika justeringar.
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
# Figure - detta är själva diagrammet. Dvs vi börjar alltid med Figure genom exempelvis fig = plt.figure() och sedan kommer de andra delarna som axes, axis och artist

fig = plt.figure() # Detta ger en tom figure.
fig, ax = plt.subplots()            

# Axes - detta är hela området som täcks av våra axlar, i detta fall x och y. 
fig, ax = plt.subplots() # detta ger en figure och en axes.

# Axis - dessa är x och y axlarna som vi kan ange värden till.
ax.plot([2, 3, 5, 10], [1, 6, 1, 8]) 

# Artist - Detta är allting som man kan se i figure, dvs. Figure, Axes och Axis.
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
# genom att använda OO-metoden.
x = np.linspace(-4, 4, 100)  # en funktion som går från -4 till 4.
y = x**2

fig, ax = plt.subplots(figsize=(5, 5), layout='constrained') # layout = constrained gör storleken optimeras. 5, 5 avgör storleken på det som plottas. 
ax.plot(x, y)  # x och y-axeln.

ax.set_xlabel('x axis')  # x-axeln.
ax.set_ylabel('y axis')  # y-axeln.
ax.set_title("OO approach")  # titel.

# genom att använda pyplot.
x = np.linspace(-4, 4, 100)  # Sample data.
y = x**2

plt.figure(figsize=(5, 5), layout='constrained')
plt.plot(x, y)  # x och y-axeln.

plt.xlabel('x axis') # x-axeln.
plt.ylabel('y axis') # y-axeln.
plt.title("pyplot approach")
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
# Skapa en figur med två subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5), layout='constrained') # ax1 och ax2 anger att det är data från två ställen vi hämtar och i subplot (1, 2) anger en rad och två kolumner för subplots.

# Första plotten - scatter plot för y = x^2
ax1.scatter(x, y) # axlarna
ax1.set_title("Scatter Plot") # namn på plotten
ax1.set_xlabel("x") # titel för x-axeln
ax1.set_ylabel("y") # titel för y-axeln

# Andra plotten - bar plot
ax2.bar(names, values) # våra listor som vi har i x och y axeln.
ax2.set_title("Bar Plot") # titeln för plotten
ax2.set_xlabel("Frukt") # x-axeln.
ax2.set_ylabel("Antal") # y-axeln.
```
